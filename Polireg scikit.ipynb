```python
from sklearn import datasets, linear_model
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import PolynomialFeatures
import numpy as np
import matplotlib.pyplot as plt
```

```python
#Se importa la base de datos de Boston

boston = datasets.load_boston()
print(boston.keys())
```

```python
X = boston.data[:,np.newaxis,5]
y = boston.target
```

```python
plt.scatter(X,y)
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size = 0.2)
```

```python
#Intancia del modelo
#Se elige el grado del polinomio
polreg = PolynomialFeatures(degree=2)
```

```python
#Se transforman los conjuntos a su forma polinomial
X_trainp = polreg.fit_transform(X_train)
X_testp = polreg.fit_transform(X_test)
```

```python
pr = linear_model.LinearRegression()
pr.fit(X_trainp,y_train)
```

```python
y_predict = pr.predict(X_testp)
```

```python
plt.scatter(X_test,y_test)
plt.plot(X_test,y_predict, color = 'g')
```

```python
print("Coeficientes: ",pr.coef_)
print("Intersección: ",pr.intercept_)
print("Precisión: ", pr.score(X_testp,y_test))
```
