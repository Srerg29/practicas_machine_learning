```python
import numpy as np
from sklearn import datasets
from sklearn.model_selection import train_test_split
from sklearn.svm import SVR
import matplotlib.pyplot as plt
```

```python
#Importar el dataset
boston = datasets.load_boston()
print(boston.data.shape)
```

```python
#Selecci√≥n de datos
X = boston.data[:,np.newaxis,5]
y = boston.target
print(X.shape)
```

```python
#Presentaci√≥n de los datos
plt.scatter(X,y)
plt.show()
```

```python
#Selecci√≥n de train y test set
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2)
```

```python
#Se define el algoritmo

#svr = SVR(C = 100.,epsilon = 0.2)
svr = SVR()
```

```python
#Entrenamiento del algoritmo
svr.fit(X_train,y_train)
```

```python
#Predicci√≥n
y_predict = svr.predict(X_test)
```

```python
plt.scatter(X, y)
plt.scatter(X_test, y_predict)
```

```python
print(svr.score(X_train,y_train))
```

```python
#Se define el algoritmo con kernel 'linear'

svrlin = SVR(kernel = 'linear',C = 100.,epsilon = 0.2)
```

```python
#Entrenamiento del algoritmo
svrlin.fit(X_train,y_train)
```

```python
#Predicci√≥n
y_predict_lin = svrlin.predict(X_test)
```

```python
plt.scatter(X, y)
plt.scatter(X_test, y_predict_lin)
```

```python
print(svrlin.score(X_train,y_train))
```
